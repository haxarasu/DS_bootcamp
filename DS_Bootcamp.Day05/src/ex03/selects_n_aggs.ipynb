```python
import pandas as pd

df = pd.read_json('../ex02/auto.json')

df.columns = df.columns.str.lower()
df.set_index('carnumber', inplace=True)

fines_filter = df['fines'] > 2100
fines_refund_filter = (df['fines'] > 2100) | (df['refund'] >= 2)
models_filter = df['model'].isin(['Focus', 'Corolla'])
numbers_filter = df[df.index.isin(['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS'])]

median_fines_by_make = df.groupby('make')['fines'].median()
median_fines_by_make_n_model = df.groupby(['make', 'model'])['fines'].median()
count_median_fines_by_make_n_model = df.groupby(['make', 'model'])['fines'].count()
min_fines_by_make_n_model = df.groupby(['make', 'model'])['fines'].min()
max_fines_by_make_n_model = df.groupby(['make', 'model'])['fines'].max()
std_deviation_fines_by_make_n_model = df.groupby(['make', 'model'])['fines'].std()

top_intruders = df.groupby(df.index)['fines'].count().sort_values(ascending=False)
top_car_number = top_intruders.index[0]
top_car_number_rows = df.loc[top_car_number]
top_intruders_by_sum = df.groupby(df.index)['fines'].sum().sort_values(ascending=False)
top_car_number_by_sum = top_intruders_by_sum.index[0]
top_car_number_by_sum_rows = df.loc[top_car_number_by_sum]

cars_with_multiple_models = df.groupby(df.index)['model'].nunique()
cars_with_multiple_models_upd = cars_with_multiple_models[cars_with_multiple_models > 1]

print("Fines Statistics:")
print("Median fines by make:")
print(median_fines_by_make)
print("\nMedian fines by make and model:")
print(median_fines_by_make_n_model)
print("\nNumber of fines by make and model:")
print(count_median_fines_by_make_n_model)
print("\nMin fines by make and model:")
print(min_fines_by_make_n_model)
print("\nMax fines by make and model:")
print(max_fines_by_make_n_model)
print("\nStandard deviation of fines by make and model:")
print(std_deviation_fines_by_make_n_model)

print("\nTop intruders by number of fines:")
print(top_intruders.head())
print("\nDetails of the most frequent violator:")
print(top_car_number_rows)

print("\nTop intruders by total fines:")
print(top_intruders_by_sum.head())
print("\nDetails of the violator with the highest total fines:")
print(top_car_number_by_sum_rows)

print("\nCars associated with multiple models:")
print(cars_with_multiple_models_upd)
```
