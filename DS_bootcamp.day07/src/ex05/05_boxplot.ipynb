```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('../../datasets/data/ab-test.csv')

control_before = data[(data['group'] == 'control') & (data['time'] == 'before')]['diff']
control_after = data[(data['group'] == 'control') & (data['time'] == 'after')]['diff']
test_before = data[(data['group'] == 'test') & (data['time'] == 'before')]['diff']
test_after = data[(data['group'] == 'test') & (data['time'] == 'after')]['diff']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

box_edge_color = 'green'
line_color = 'red'

bp1 = ax1.boxplot([control_before, test_before],
                 labels=['Control', 'Test'],
                 boxprops=dict(linewidth=3, color=box_edge_color),
                 whiskerprops=dict(color=line_color),
                 capprops=dict(color=line_color),
                 medianprops=dict(linewidth=2, color=line_color),
                 patch_artist=True)

for box in bp1['boxes']:
    box.set_facecolor('none')

ax1.set_title('Before', fontsize=15)
ax1.grid(axis='both')

bp2 = ax2.boxplot([control_after, test_after],
                 labels=['Control', 'Test'],
                 boxprops=dict(linewidth=3, color=box_edge_color),
                 whiskerprops=dict(color=line_color),
                 capprops=dict(color=line_color),
                 medianprops=dict(linewidth=2, color=line_color),
                 patch_artist=True)

for box in bp2['boxes']:
    box.set_facecolor('none')

ax2.set_title('After', fontsize=15)
ax2.grid(axis='both')

fig.suptitle('Lag before labs deadlines. After and before the Newsfeed', fontsize=15)

plt.tight_layout()
plt.show()
```
