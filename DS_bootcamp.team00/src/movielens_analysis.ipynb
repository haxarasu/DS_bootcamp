```python
%%timeit

from classes.links import Links
from classes.movies import Movies
from classes.ratings import Ratings
from classes.tags import Tags

links_result = Links("../ml-latest-small/links.csv")
movies_result = Movies("../ml-latest-small/movies.csv")
ratings_result = Ratings("../ml-latest-small/ratings.csv")
users_result = ratings_result.users
tags_result = Tags("../ml-latest-small/tags.csv")

report = []

report.append("###Отчет по модулю links###\n")
report.append("\n1. Список списков с информацией с веб-страниц:\n")
report.append(links_result.get_imdb([25, 37, 43], ["title", "year", "genre"]))
report.append("\n2. Топ-10 режиссерами по количеству фильмов\n")
report.append(links_result.top_directors(10))
report.append("\n3. Топ-10 фильмов по величине бюджета:\n")
report.append(links_result.most_expensive(10))
report.append("\n4. Топ-10 фильмов по величине прибыли:\n")
report.append(links_result.most_profitable(10))
report.append("\n5. Топ-10 фильмов по продолжительности:\n")
report.append(links_result.longest(10))
report.append("\n6. Топ-10 фильмов по стоимости за минуту:\n")
report.append(links_result.top_cost_per_minute(10))

report.append("###Отчет по модулю movies###\n")
report.append("\n1. Распределение фильмов по годам выпуска:\n")
report.append(movies_result.dist_by_release())
report.append("\n2. Самые популярные жанры:\n")
report.append(movies_result.dist_by_genres())
report.append("\n3. Топ-10 фильмов с самым большим количеством жанров:\n")
report.append(movies_result.most_genres(10))

report.append("###Отчет по модулю ratings###\n")
report.append("\n1. Годы выпуска фильмов по возрастанию:\n")
report.append(ratings_result.movies.dist_by_year())
report.append("\n2. Распределение рейтингов для фильмов:\n")
report.append(ratings_result.movies.dist_by_rating())
report.append("\n3. Топ-10 фильмов с самым большим количеством оценок:\n")
report.append(ratings_result.movies.top_by_num_of_ratings(10))
report.append("\n4. Топ-10 фильмов с самым большим средним рейтингом:\n")
report.append(ratings_result.movies.top_by_ratings(10, metric="average"))  
report.append("\n5. Топ-10 фильмов с самым большим медианным рейтингом:\n")
report.append(ratings_result.movies.top_by_ratings(10, metric="median"))  
report.append("\n6. Топ-10 фильмов с самыми противоречивыми оценками:\n")
report.append(ratings_result.movies.top_controversial(10))
report.append("\n7. Топ-10 пользователей с самым большим количеством оценок:\n")
report.append(users_result.top_by_num_of_ratings(10))
report.append("\n8. Топ-10 пользователей с самой большой средней оценкой:\n")
report.append(users_result.top_by_ratings(10, metric="average"))  
report.append("\n9. Топ-10 пользователей с самой большой медианой оценки:\n")
report.append(users_result.top_by_ratings(10, metric="median"))
report.append("\n10. Топ-10 пользователей с самыми противоречивыми оценками:\n")
report.append(users_result.top_controversial(10))

report.append("###Отчет по модулю tags###\n")
report.append("\n1. Топ-10 тегов по количеству слов:\n")
report.append(tags_result.most_words(10))
report.append("\n2. Топ-10 тегов по количеству символов:\n")
report.append(tags_result.longest(10))
report.append("\n3. Топ-10 тегов с самым длинным названием по количеству символов и слов:\n")
report.append(tags_result.most_words_and_longest(10))
report.append("\n4. Топ-10 самых популярных тегов:\n")
report.append(tags_result.most_popular(10))
report.append("\n5. Теги со словом 'cool':\n")
report.append(tags_result.tags_with('cool'))

with open("report.txt", "w") as f:
    for item in report:
        f.write(str(item) + "\n")
```
